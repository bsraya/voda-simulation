```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bb = pd.read_csv('bb.csv')
km = pd.read_csv('munkres.csv')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# plot km and bb
ax[0].plot(km['nodes'], km['duration(ms)'], 'r', label='Kuhn-Munkres')
ax[0].plot(bb['nodes'], bb['duration(ms)'], 'b', label='Branch & Bound')

# show legend
ax[0].legend()

# show x and y labels
ax[0].set_title('Job Makespans')
ax[0].set_xlabel('Nodes')
ax[0].set_ylabel('Microseconds')

# plot km and bb
ax[1].plot(km['nodes'], km['cost'], 'r', label='Kuhn-Munkres')
ax[1].plot(bb['nodes'], bb['cost'], 'b', label='Branch & Bound')

# show legend
ax[1].legend()

# show x and y labels
ax[1].set_title('Job Cost')
ax[1].set_xlabel('Nodes')
ax[1].set_ylabel('Costs')
```

```python
bb_gantt = pd.read_csv('2nodes_bb_gantt.csv')

fig, ax = plt.subplots(1, figsize=(16,6))
ax.barh(bb_gantt['job'], bb_gantt['running(s)'], left=bb_gantt['waiting(s)'], color=['#46244C', '#46200C', '#712B75', '#C74B00', '#C74B50', '#D20B50', '#D49B54', '#505050', '#ffa500', '#D00B00'])
ax.invert_yaxis()
plt.title('Branch & Bound Gantt Chart')
plt.xlabel('Time (s)')
plt.show()
```

```python
avg_waiting_time = sum(bb_gantt['waiting(s)'])/len(bb_gantt['waiting(s)'])
print('Average waiting time: ' + str(avg_waiting_time) + ' seconds')
```
